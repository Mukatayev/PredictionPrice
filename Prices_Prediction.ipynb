```python
# Delete the less corr variable to the price (<0.5)
data.drop(["id",
           "date",
           "bedrooms",
           "sqft_lot","floors",
           "waterfront","view",
           "condition",
           "sqft_basement",
           "yr_built",
           "yr_renovated",
           "zipcode",
           "lat","long",
           "sqft_lot15"],axis=1, inplace=True)
```

```python
# Reg eq y = mx + c
y = data["price"]
x = data.drop("price",axis=1)
x
```

```python
# Train and Test 
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split (x, y, test_size = 0.33, random_state = 100)

#Building Linar regression Models
from sklearn.linear_model import LinearRegression

#Linear Regression
reg_model = LinearRegression()
reg_model.fit(xtrain, ytrain)

# Prediction
prediction = reg_model.predict(xtest)
```

```python
plt.xlabel("ytest")
plt.ylabel("predicted values")
plt.scatter(ytest,prediction)
```

```python
diff_show = pd.DataFrame({"Actual":ytest, "Predicted" : prediction})
diff_show
```

```python
# R square 
from sklearn.metrics import r2_score
r2_score(ytest,prediction)
```

```python
# 54% rel btw model and depend
```

```python
# Model Reg Simple
data = pd.read_csv("C:/Users/Adila/Desktop/Data/kc_house_data.csv")
data = pd.DataFrame(data)

y = data["price"]
x = data["sqft_living"]
x.dtypes
```

```python
x
```

```python
x = np.array(x).reshape(-1, 1)
```

```python
y = np.array(y)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.4, random_state=50)
```

```python
model3 = LinearRegression()
model3.fit(xtrain, ytrain)

prediction = model3.predict(xtest)
```

```python
# Training test
plt.scatter(xtrain, ytrain)
plt.plot(xtrain, model3.predict(xtrain), color = 'red')
plt.show()
```

```python
#Test Results 
plt.scatter(xtest, ytest)
plt.plot(xtrain, model3.predict(xtrain), color = 'red')
plt.show()
```

```python
r2_score(ytest, prediction)
```
